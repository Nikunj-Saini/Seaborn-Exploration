```python
import  seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# 1. Scatter plot using relplot

```python
sns.relplot(data = tips , x = "total_bill", y = "tip" , kind = "scatter")
plt.suptitle("Relplot - Scatter Example")
plt.show()
```

# 2. Scatter with hue and size

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", size="size", kind="scatter")
plt.suptitle("Relplot - Line Example ") 
plt.show()
```

# 3. Line plot using relplot

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line", hue="region", style="event")
plt.suptitle("Relplot - Line Example")
plt.show()
```

# 4. FacetGrid with relplot (row & col)

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="time", row="smoker", kind="scatter")
plt.suptitle("Relplot with FacetGrid (Row & Col)")
plt.show()
```
