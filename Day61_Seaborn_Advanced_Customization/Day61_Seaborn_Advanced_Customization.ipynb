```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# 1. Seaborn plot + Matplotlib Annotation

```python
# 1. Seaborn plot + Matplotlib annotation
plt.figure(figsize=(7,5))
ax = sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Tips vs Total Bill")

# Annotating Highest Tip Point
max_tip_row = tips.loc[tips['tip'].idxmax()]
plt.annotate(f"Highest Tip: {max_tip_row['tip']}",
             xy=(max_tip_row['total_bill'], max_tip_row['tip']),
             xytext=(max_tip_row['total_bill']+5, max_tip_row['tip']+1),
             arrowprops=dict(arrowstyle="->", color='red'))
plt.show()
```

# 2. Multiple Axes Example

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))
sns.histplot(tips['total_bill'], bins=20, ax=ax[0], color='skyblue')
ax[0].set_title("Total Bill Distribution")
sns.histplot(tips['tip'], bins=20, ax=ax[1], color='orange')
ax[1].set_title("Tip Distribution")
plt.tight_layout()
plt.show()
```

# 3. Twin Axes Example

```python
fig, ax1 = plt.subplots(figsize=(7,5))
sns.lineplot(x='size', y='tip', data=tips, ax=ax1, color='blue', label='Tip')
ax2 = ax1.twinx()
sns.lineplot(x='size', y='total_bill', data=tips, ax=ax2, color='green', label='Total Bill')

ax1.set_xlabel("Party Size")
ax1.set_ylabel("Tip", color='blue')
ax2.set_ylabel("Total Bill", color='green')
ax1.set_title("Tip vs Total Bill by Party Size")

# handle legends manually
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1+lines2, labels1+labels2, loc='upper left')
plt.show()
```
