```python
import seaborn as sns
import matplotlib.pyplot as plt 
tips = sns.load_dataset('tips')
```

# 1. Basic Violin Plot

```python
plt.figure(figsize= (8,5))
sns.violinplot(x = "day", y = "total_bill", data = tips , palette = "Set2")
plt.title("Basic Violin Plot  -  Total Bill by Day ")
plt.show()
```

# 2. Violin Plot with Hue (split categories)

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2", split=True)
plt.title("Violin Plot with Hue (Split by Sex)")
plt.show()
```

# 3. Violin Plot with Quartiles`

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="total_bill", data=tips, inner="quartile", palette="pastel")
plt.title("Violin Plot with Quartiles")
plt.show()
```

# 4. Horizontal Violin Plot

```python
plt.figure(figsize=(8,6))
sns.violinplot(y="day", x="total_bill", data=tips, palette="muted")
plt.title("Horizontal Violin Plot")
plt.show()
```

# 5. Violin Plot + Swarmplot Overlay

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set3", inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black", alpha=0.6)
plt.title("Violin Plot with Swarmplot Overlay")
plt.show()
```
