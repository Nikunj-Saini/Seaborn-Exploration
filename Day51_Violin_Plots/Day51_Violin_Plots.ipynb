```python
import seaborn as sns 
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# 1. Basic violin plot

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Basic Violin Plot")
plt.show()
```

# 2. Violin plot with hue (split by gender)

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)
plt.title("Violin Plot with Split by Gender")
plt.show()
```

# 3. Violin plot without inner markings

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
plt.title("Violin Plot without Inner Markings")
plt.show()
```

# 4. Violin plot with box inside

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner="box")
plt.title("Violin Plot with Box Inside")
plt.show()
```

# 5. Control width by number of observations

```python
sns.violinplot(x="day", y="total_bill", data=tips, scale="count")
plt.title("Violin Plot with Scale by Count")
plt.show()
```
