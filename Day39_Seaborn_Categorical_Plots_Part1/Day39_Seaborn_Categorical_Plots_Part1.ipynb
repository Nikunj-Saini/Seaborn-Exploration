# BAR PLOT 

## It helps us to show the relation between a categorical variable (x-axis) and a Numerical varialbe (y - axis )

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips, palette="Set2")

plt.title("Average Total Bill by Day", fontsize=16)

plt.show()
```

# 2.  COUNT PLOT 

--It is used to show counts of observations in each category.

--It is Similar to a histogram but for a categorical data.

```python
plt.figure(figsize=(8,6))

sns.countplot(x="day", data=tips, palette="Set3")
plt.title("Number of Observations by Day", fontsize=16)

plt.show()
```

# 3. Adding Hue

```python
plt.figure(figsize=(8,6))

sns.barplot(x="day", y="total_bill", hue="sex", data=tips, palette="pastel")
plt.title("Average Total Bill by Day & Gender", fontsize=16)

plt.show()


plt.figure(figsize=(8,6))

sns.countplot(x="day", hue="sex", data=tips, palette="muted")
plt.title("Number of Observations by Day & Gender", fontsize=16)

plt.show()
```
