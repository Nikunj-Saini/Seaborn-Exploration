```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

# 1. Default Plot

```python
sns.scatterplot(x = "total_bill" , y = "tip", hue = "sex" , data = tips )
plt.title("Default Style")
plt.show()
```

# 2. Change Style

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Whitegrid Style")
plt.show()
```

# 3. Change Context

```python
sns.set_context("talk")  # larger elements
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Whitegrid + Talk Context")
plt.show()
```

# 4. Change Color Palette

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.set_palette("muted")
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Muted Palette")
plt.show()
```

# 5. Custom Palette Example

```python
custom_palette = sns.color_palette("coolwarm", as_cmap=False)
sns.set_palette(custom_palette)
plt.figure(figsize=(6,4))
sns.histplot(tips["total_bill"], bins=20)
plt.title("Coolwarm Palette on Histogram")
plt.show()
```

# 6. Reset to Defaults

```python
sns.reset_defaults()
```
