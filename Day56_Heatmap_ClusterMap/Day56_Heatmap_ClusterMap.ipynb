```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Basic heatmap
sns.heatmap(flights_pivot)
plt.title("Flight Passengers Heatmap")
plt.show()
```

# 2. Add Annotations & Styling

```python
sns.heatmap(
    flights_pivot,annot=True,fmt="d",cmap="YlGnBu",linewidths=.5,       
    cbar_kws={"label": "Passengers"}
)
plt.title("Flight Passengers Heatmap with Annotations")
plt.show()
```

# 3. ClusterMap

```python
sns.clustermap(flights_pivot,cmap="coolwarm",standard_scale=1,annot=True,fmt="d")
plt.show()
```
