```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

# 1. Basic Strip Plot

```python
plt.figure(figsize = (8,5))
sns.stripplot(x = "day", y = "total_bill", data = tips )
plt.title("Basic Strip Plot : Total Bill by Day")
plt.show()
```

# 2. Strip Plot with Jitter

```python
plt.figure(figsize=(8,6))
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, palette="Set2")
plt.title("Strip Plot with Jitter")
plt.show()
```

# 3. Strip Plot with Hue

```python
plt.figure(figsize=(8,6))
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex", palette="Set1", jitter=True, dodge=True)
plt.title("Strip Plot with Hue (Grouped by Sex)")
plt.show()
```

# 4. Horizontal Strip Plot

```python
plt.figure(figsize=(8,6))
sns.stripplot(y="day", x="total_bill", data=tips, jitter=True, palette="muted")
plt.title("Horizontal Strip Plot")
plt.show()
```

# 5. Strip Plot + Boxplot

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
sns.stripplot(x="day", y="total_bill", data=tips, color="black", alpha=0.5)
plt.title("Strip Plot Overlay with Boxplot")
plt.show()
```
