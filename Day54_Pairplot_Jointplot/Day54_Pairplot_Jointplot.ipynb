# 1. Pairplot â€“ Pairwise Relationships 

```python
import seaborn as sns 
import matplotlib.pyplot as plp 

iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

# 2. Pairplot with Hue (Categorical Coloring)

```python
sns.pairplot(iris, hue="species", palette="Set2")
plt.show()
```

# 3. Jointplot â€“ Focus on Two Variables

```python
sns.jointplot(data = iris, x = "sepal_length" , y = "sepal_width")
plt.show()
```

# 4. Jointplot with Different Kinds

```python
# Scatter + regression line
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind="reg")

# Hexbin style
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind="hex")
plt.show()
```
