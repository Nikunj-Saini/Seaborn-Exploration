# Relational Plots

## 1. Scatter Plot (sns.scatterplot)

--It is best for showing relationship between Two numerical variables 

--Example  Age  vs Salary, or Total Bill vs Tip

--With this we can add colors for  Different Categories , style (marker) for distinction , and Size for an additional dimention 

```python
import seaborn as sns 
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

plt.figure(figsize=(8,4))
sns.scatterplot(data = tips, x = "total_bill",y = "tip", hue = "sex",
                style = "smoker" , size = "size", palette = "Set2", s = 100)
plt.title("scatterplot : Total Bill VS Tip")
plt.show()
```

## 2. Line Plot (sns.lineplot)

--Great for trends over time or countinuous variables 

--Automatically computes confidence interval

```python
plt.figure(figsize=(8,6))
sns.lineplot(data = tips , x = "size", y = "tip", hue = "sex" , 
             style = "smoker" , markers = True , dashes  =False)
plt.title("Lineplot: Tip by Table Size ")
plt.show()
```
