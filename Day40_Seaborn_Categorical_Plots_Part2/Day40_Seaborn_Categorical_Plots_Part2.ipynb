# 1. BOXPLOT

--It is usefull for identifying Spread & Outlier 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.boxplot(x = "day", y = "total_bill", data = tips, palette = "Set2")
plt.show()
```

# 2. VIOLIN PLOT 

--It combines boxplot  = KDE 

--It shows both the Shape and Summary

```python
sns.violinplot(x = "day" , y = "total_bill", data = tips , palette = "muted")
plt.title(" Violin plot of Total Bill by the day")
plt.show()
```

# 3. STRIP PLOT 

```python
sns.stripplot(x = "day" , y = "total_bill", data = tips , palette = "Set3")
plt.title(" Strip Plot of Total Bill by the Day")
plt.show()
```

# 4. SWARM PLOT 

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Swarmplot of Total Bill by Day", fontsize=16)
plt.show()
```

# 5.Combining Plots

--It can overlay stripplot on top of violinplot for richer visualization.

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="total_bill", data=tips, inner=None, palette="muted")
sns.stripplot(x="day", y="total_bill", data=tips, color="black", size=3)
plt.title("Violin + Stripplot", fontsize=16)
plt.show()
```
