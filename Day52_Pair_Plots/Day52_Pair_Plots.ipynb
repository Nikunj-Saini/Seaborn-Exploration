```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
```

# 1. Basic pairplot

```python
sns.pairplot(iris)
plt.show()
```

# 2. Pairplot with hue for species

```python
sns.pairplot(iris, hue="species")
plt.show()
```

# 3. Pairplot with only selected columns

```python
sns.pairplot(iris, vars=["sepal_length", "sepal_width"], hue="species")
plt.show()
```

# 4. Pairplot with kind="kde" (density plots)

```python
sns.pairplot(iris, hue="species", kind="kde")
plt.show()
```

# 5. Pairplot with diagonal kind as histogram

```python
sns.pairplot(iris, hue="species", diag_kind="hist")
plt.show()
```
