```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

# 1. Basic histogram with displot

```python
sns.displot(data=tips, x="total_bill")
plt.suptitle("Displot - Basic Histogram")
plt.show()
```

# 2. Histogram with hue

```python
sns.displot(data=tips, x="total_bill", hue="sex", multiple="stack")
plt.suptitle("Displot - Histogram with Hue (Stacked)")
plt.show()
```

# 3. Histogram with KDE

```python
sns.displot(data=tips, x="total_bill", kde=True, bins=20)
plt.suptitle("Displot - Histogram with KDE")
plt.show()
```

# 4. FacetGrid with displot

```python
sns.displot(data=tips, x="total_bill", col="time", row="smoker", kde=True)
plt.suptitle("Displot with FacetGrid")
plt.show()
```
