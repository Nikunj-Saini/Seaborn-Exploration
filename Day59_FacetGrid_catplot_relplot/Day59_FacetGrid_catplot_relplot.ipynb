# 1. FacetGrid â€“ Full Control

```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

g = sns.FacetGrid(tips, col="sex",row = "smoker")
g.map(sns.scatterplot, "total_bill", "tip") 
plt.show()
```

# 2. catplot â€“ High-Level Categorical Grid]

```python
sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex", col="smoker", kind = "box")
plt.show()
```

# 3. Relplot â€“ High-Level Relational Grid

```python
penguins = sns.load_dataset("penguins")
sns.relplot(data = penguins,x="flipper_length_mm", y="body_mass_g",hue="species",col="sex",kind="scatter")
plt.show() 
```
