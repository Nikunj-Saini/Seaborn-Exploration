# 1 Basic Swarm Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarm Plot of Total Bill by Day", fontsize=16)
plt.show()
```

# 2 Adding Hue

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Swarm Plot with Hue (Sex)", fontsize=16)
plt.show()
```

# 3 Customizations (dodge + palette + alpha)

```python
plt.figure(figsize=(8,6))
sns.swarmplot(    x="day", y="total_bill", hue="sex", data=tips, dodge=True, palette="Set2", alpha=0.8)
plt.title("Customized Swarm Plot", fontsize=16)
plt.show()
```
