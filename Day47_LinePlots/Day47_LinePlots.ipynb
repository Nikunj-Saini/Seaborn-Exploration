```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
```

# 1. Basic line plot

```python
sns.lineplot(data=fmri, x="timepoint", y="signal")
plt.title("Basic Line Plot")
plt.show()
```

# 2. Line plot with hue

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
plt.title("Line Plot with Hue (Event)")
plt.show()
```

# 3. Line plot with style

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot with Style (Region)")
plt.show()
```

# 4. Line plot with multiple dimensions

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event", size="subject")
plt.title("Line Plot with Hue, Style, and Size")
plt.show()
```
