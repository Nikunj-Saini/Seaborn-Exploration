```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# 1. Basic stripplot

```python
sns.stripplot(x = "day" , y = "total_bill" , data = tips , jitter = True)
plt.title("Strip plot with Jitter ")
plt.show()
```

# 2. Stripplot with hue

```python
sns.swarmplot( x= "day" , y = "total_bill" ,  data = tips)
plt.title("Swarm plot  Avoids Overlap")
plt.show()
```

# 3. Swarmplot (better for overlap)

```python
sns.swarmplot(x = "day", y = "total_bill" , data = tips , hue = "sex", dodge = True)
plt.title("swarm plot with hue and Dodge")
plt.show()
```

# 4. Swarmplot with hue

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
plt.title("Swarm Plot with Hue and Dodge")
plt.show()
```
