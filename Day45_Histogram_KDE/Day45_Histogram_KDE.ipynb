```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# 1. Simple Histogram of total_bill

```python
sns.histplot(data = tips, x = "total_bill", bins = 20, color = "skyblue", edgecolor = "black")
plt.title("Histogram of Total Bill ")
plt.show()
```

# 2. Histogram with KDE curve

```python
sns.histplot(data = tips , x = "total_bill", kde = True , color = "lightcoral")
plt.title("Distribution of Total Bill by Gender")
plt.show()
```

# 3. Compare Distributions by 'sex'

```python
sns.histplot(data=tips, x="total_bill", hue="sex", kde=True, multiple="stack", palette="Set2")
plt.title("Distribution of Total Bill by Gender")
plt.show()
```

# 4. Multiple Distributions Overlapping

```python
sns.histplot(data=tips, x="total_bill", hue="smoker", kde=True, element="step", stat="density")
plt.title("Histogram & KDE - Smokers vs Non-Smokers")
plt.show()
```
