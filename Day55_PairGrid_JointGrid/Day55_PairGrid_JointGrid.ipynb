# 1. PairGrid â€“ Custom Pairwise Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

g = sns.PairGrid(iris, hue="species", palette="Set1")

# Upper triangle
g.map_upper(sns.scatterplot)

# lower triangle
g.map_lower(sns.kdeplot, fill=True) 

# Diagonal
g.map_diag(sns.histplot)      

g.add_legend()
plt.show()
```

# 2. JointGrid â€“ Custom Joint Plots

```python
g = sns.JointGrid(data = iris, x = "sepal_length", y = "sepal_width")

g.plot(sns.scatterplot, sns.histplot)

sns.regplot(data = iris , x = "sepal_length" , y = "sepal_width",ax=g.ax_joint, scatter = False )
plt.show()
```
