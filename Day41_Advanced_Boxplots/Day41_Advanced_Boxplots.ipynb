# 1. Basic Horizontal Boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
plt.figure(figsize= (8,5))
sns.boxplot(x = "total_bill", y = "day", data = tips, palette = "Set2")
plt.title("Horizontal Boxplot of Total Bill by Day")
plt.show()
```

# 2. Adding Swarmplot with Boxplot

```python
plt.figure (figsize=(8,5))
sns.boxplot(x = "day" , y = "total_bill", data = tips, palette = "Set2")
sns.swarmplot(x = "day", y = "total_bill", data = tips, color = "black", alpha = 0.7)
plt.title("Boxplot with Swarmplot Overlay")
plt.show()
```

# 3. Grouping with hue

```python
plt.figure(figsize= (8,5))
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = tips, palette = "Set2")
plt.title("Boxplot by Day and Sex")
plt.show()
```

# 4. Customization (linewidth, width, dodge)

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set3",
            linewidth=2, width=0.6, dodge=True)
plt.title("Customized Boxplot (Smoker vs Non-Smoker)")
plt.show()
```
