```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
```

# BAR PLOT 

```python
plt.figure(figsize = (6,4))
sns.barplot(x = "sex", y ="tip", data = tips , palette = "pastel")
plt.title("Average Tip By gender ")
plt.show()
```

# COUNT PLOT 

```python
plt.figure(figsize=(6,4))
sns.countplot(x="smoker", data=tips, palette="muted")
plt.title("Smokers vs Non-Smokers")
plt.show()
```

# BOX PLOT 

```python
plt.figure(figsize=(6,4))
sns.boxplot(x = "day", y = "total_bill" , data = tips, palette="Set2")
plt.title("Box Plot of Total Bill  by Day")
plt.show()
```

# VIOLIN PLOT 

```python
plt.figure(figsize = (7,4))
sns.violinplot(x = "sex", y = "tip", data = tips , palette = "coolwarm")
plt.title("Violin Plot of Tips by Gender ")
plt.show()
```
