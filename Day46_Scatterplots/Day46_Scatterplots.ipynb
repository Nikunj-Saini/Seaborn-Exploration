```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

# 1. Basic Scatterplot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Basic Scatterplot: Total Bill vs Tip")
plt.show()
```

# 2. Scatterplot with hue (color by category)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette="Set2")
plt.title("Scatterplot with Hue (Sex)")
plt.show()
```

# 3. Scatterplot with size

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size", hue="sex", palette="Set2")
plt.title("Scatterplot with Size and Hue")
plt.show()
```

# 4. Scatterplot with style (different marker shapes)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker", palette="Set1")
plt.title("Scatterplot with Style by Smoker Status")
plt.show()
```
